En este ejercicio, formar谩s dos grupos aleatorios de estudiantes, y cada grupo explorar谩 uno de los conceptos fundamentales en Python: mapa, reducir o filtrar. Tu tarea es comprender estos conceptos a fondo y preparar una explicaci贸n clara para compartir con la clase.

````python
import random
from typing import List, Tuple

def assign_students_to_groups(students: List[str], roles: List[str], num_groups: int) -> List[List[Tuple[str, str]]]:
    """
    Assign students to groups randomly, each with a specified role.

    Args:
    - students (List[str]): A list of student names.
    - roles (List[str]): A list of roles to assign to students.
    - num_groups (int): The number of groups to create.

    Returns:
    - List[List[Tuple[str, str]]]: A list of groups, where each group is a list of tuples.
      Each tuple contains a student's name and their assigned role.
    """
    # Shuffle the list of students randomly
    random.shuffle(students)
    
    # Divide the students into groups
    groups = [[] for _ in range(num_groups)]
    
    for i, student in enumerate(students):
        group_index = i % num_groups
        role = roles[i % len(roles)]
        groups[group_index].append((student, role))
    
    return groups
````

````python
# List of students
students = ["Sara", "Glennys", "Alex", "Adr铆an", "Andr茅s", "Marta"]

# List of roles
roles = ["map", "filter", "reduce"]

# Number of groups
num_groups = 2

# Assign students to groups
groups = assign_students_to_groups(students, roles, num_groups)

# Print the groups
for i, group in enumerate(groups, start=1):
    print(f"Group {i}: {group}")
````

锔 Reducir es un poco m谩s complicado de entender que map() y filter(), as铆 que vamos a seguir el siguiente ejemplo 锔

````python
from functools import reduce

# Definir una lista de n煤meros para la cual queremos calcular el producto.
n煤meros = [1, 2, 3, 4, 5]

# Crear una funci贸n que toma dos argumentos y devuelve su producto.
# Usaremos esta funci贸n como el "acumulador" en reduce().
# La funci贸n reduce() aplicar谩 esta funci贸n acumulativamente a los elementos de la lista.
def multiplicar(x, y):
    return x * y

# Usar la funci贸n reduce() para calcular el producto de todos los n煤meros en la lista.
# Pasar la funci贸n multiplicar y la lista de n煤meros como argumentos a reduce().
producto = reduce(multiplicar, n煤meros)

# Imprimir el resultado
print(producto)  # Salida: 120
````

#### 驴Qu茅 sucedi贸?
- Comenzamos con una lista [1, 2, 3, 4, 5] y pasamos la funci贸n multiply(x, y) a reduce() junto con esta lista, sin un valor inicial.

- reduce() llama a multiply(1, 2), y multiply() devuelve 2.

- Toma el resultado (2) y el siguiente elemento de la lista (3).

- Aplica la funci贸n multiply de nuevo: 2 * 3 = 6.

- Contin煤a este proceso hasta que todos los elementos en la lista hayan sido procesados.

Dado que no quedan m谩s elementos en la secuencia, reduce() devuelve 16.
